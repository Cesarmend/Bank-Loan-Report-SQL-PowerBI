```python
import pandas as pd
```

```python
df=pd.read_csv('financial_loan.csv',sep=',',encoding='utf-8')
```

```python
# List of date columns to check for null values
date_columns = ['issue_date', 'last_credit_pull_date', 'last_payment_date', 'next_payment_date']

# Check for null values in the specified date columns
null_dates = df[date_columns].isnull()

# Display rows where any of the date columns have null values
null_rows = df[null_dates.any(axis=1)]

print("Rows with null values in date columns:")
print(null_rows)
```

```python
from dateutil import parser

def convert_to_standard_date(date_str):
    try:
        # Parse the date using dateutil.parser with dayfirst=True to prioritize DD/MM/YYYY formats
        parsed_date = parser.parse(date_str, dayfirst=True)
        # Format the parsed date to the preferred format (YYYY-MM-DD)
        return parsed_date.strftime('%Y-%m-%d')
    except (ValueError, TypeError):
        # Return the original value if it cannot be parsed as a date
        return date_str

# Specify the columns in your existing DataFrame that contain dates
date_columns = ['issue_date', 'last_credit_pull_date', 'last_payment_date', 'next_payment_date']

# Apply the conversion to the specified date columns in-place
for col in date_columns:
    df[col] = df[col].apply(convert_to_standard_date)
```

```python
# List of date columns to check for null values
date_columns = ['issue_date', 'last_credit_pull_date', 'last_payment_date', 'next_payment_date']

# Check for null values in the specified date columns
null_dates = df[date_columns].isnull()

# Display rows where any of the date columns have null values
null_rows = df[null_dates.any(axis=1)]

print("Rows with null values in date columns:")
print(null_rows)
```
